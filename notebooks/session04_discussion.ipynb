# 導入: Abstract で研究全体を再確認する

前回のイントロダクションと同じ論文を教材に使います。

論文リンク: [Intra-axial primary brain tumor differentiation: comparing large language models on structured MRI reports vs. radiologists on images](https://link.springer.com/article/10.1007/s00330-025-11924-3)

## Abstract

### Objective

> Aimed to evaluate the potential of large language models (LLMs) in differentiating intra-axial primary brain tumors using structured magnetic resonance imaging (MRI) reports and compare their performance with radiologists.

日本語訳: 構造化された magnetic resonance imaging (MRI) レポートを用いて、脳実質内原発性脳腫瘍を鑑別する際の large language models (LLMs) の可能性を評価し、その性能を放射線科医と比較することを目的とした。

### Materials and methods

> Structured reports of preoperative MRI findings from 137 surgically confirmed intra-axial primary brain tumors, including Glioblastoma (n = 77), Central Nervous System (CNS) Lymphoma (n = 22), Astrocytoma (n = 9), Oligodendroglioma (n = 9), and others (n = 20), were analyzed by multiple LLMs, including GPT-4, Claude-3-Opus, Claude-3-Sonnet, GPT-3.5, Llama-2-70B, Qwen1.5-72B, and Gemini-Pro-1.0. The models provided the top 5 differential diagnoses based on the preoperative MRI findings, and their top 1, 3, and 5 accuracies were compared with board-certified neuroradiologists’ interpretations of the actual preoperative MRI images.

日本語訳: 手術により確認された 137 例の脳実質内原発性脳腫瘍について、術前 MRI 所見の構造化レポートを解析対象とした。内訳は Glioblastoma (n = 77)、Central Nervous System (CNS) Lymphoma (n = 22)、Astrocytoma (n = 9)、Oligodendroglioma (n = 9)、その他 (n = 20) であった。これらのレポートを、GPT-4、Claude-3-Opus、Claude-3-Sonnet、GPT-3.5、Llama-2-70B、Qwen1.5-72B、Gemini-Pro-1.0 を含む複数の LLM で解析した。各モデルは術前 MRI 所見に基づいて上位 5 つの鑑別診断を提示し、その top 1、top 3、top 5 accuracy を、実際の術前 MRI 画像を読影した専門医資格を有する神経放射線科医の解釈と比較した。

### Results

> Radiologists achieved top 1, 3, and 5 accuracies of 85.4%, 94.9%, and 94.9%, respectively. Among the LLMs, GPT-4 performed best with top 1, 3, and 5 accuracies of 65.7%, 84.7%, and 90.5%, respectively. Notably, GPT-4’s top 3 accuracy of 84.7% approached the radiologists’ top 1 accuracy of 85.4%. Other LLMs showed varying performance levels, with average accuracies ranging from 62.3% to 75.9%. LLMs demonstrated high accuracy for Glioblastoma but struggled with CNS Lymphoma and other less common tumors, particularly in top 1 accuracy.

日本語訳: 放射線科医の top 1、top 3、top 5 accuracy は、それぞれ 85.4%、94.9%、94.9% であった。LLM の中では GPT-4 が最も高い性能を示し、top 1、top 3、top 5 accuracy はそれぞれ 65.7%、84.7%、90.5% であった。特に、GPT-4 の top 3 accuracy である 84.7% は、放射線科医の top 1 accuracy である 85.4% に近かった。他の LLM はさまざまな性能を示し、平均 accuracy は 62.3% から 75.9% の範囲であった。LLM は Glioblastoma に対して高い accuracy を示したが、CNS Lymphoma やその他の頻度の低い腫瘍では苦戦し、特に top 1 accuracy でその傾向がみられた。

### Conclusion

> LLMs show promise as assistive tools for differentiating intra-axial primary brain tumors using structured MRI reports. However, a significant gap remains between their performance and that of board-certified neuroradiologists interpreting actual images. The choice of LLM and tumor type significantly influences the results.

日本語訳: LLM は、構造化 MRI レポートを用いて脳実質内原発性脳腫瘍を鑑別するための支援ツールとして有望である。しかし、LLM の性能と、実際の画像を読影する専門医資格を有する神経放射線科医の性能との間には、依然として大きな差がある。使用する LLM の種類と腫瘍の種類は、結果に大きく影響する。

# Part 1. Discussion とは

Discussion は、Results で示した結果をもとに、研究全体の意味を説明するセクションです。

Discussion で書くことは、主に次の5つです。

1. 本研究の主要結果
2. 結果の解釈
3. 先行研究との比較
4. 臨床的・科学的意義
5. 研究の限界と結論

重要なのは、Discussion は「感想を書く場所」ではないということです。Results で得られた事実をもとに、論理的に意味づけを行います。


# Part 2. Results と Discussion の違い

| セクション | 書くこと | 書かないこと |
|---|---|---|
| Results | 得られた結果、数値、表・図の説明 | 深い解釈、理由、臨床的意味 |
| Discussion | 結果の意味、先行研究との関係、限界、結論 | Results にない新しい解析や数値 |

例:

```markdown
Results: GPT-4 achieved the highest top-1 accuracy among the LLMs.

Discussion: This finding suggests that larger or more advanced LLMs may be better able to use structured imaging findings to generate plausible differential diagnoses.
```

Results は「何が起きたか」、Discussion は「それは何を意味するか」を書く部分です。

> Despite these challenges, recent LLMs demonstrate considerable potential as assistive tools for generating differential diagnoses in our study. The stark contrast in Top 1 diagnostic accuracy for brain tumors, particularly for less common types such as CNS lymphoma, highlights the challenges LLMs face in achieving the level of expertise exhibited by board-certified radiologists. These limitations may stem from the probabilistic nature of LLM training… As a result, while LLMs can assist in generating differential diagnoses, they are less likely to prioritize rare diseases as the top diagnosis. It is also plausible that the observed performance difference between common tumors like Glioblastoma and less common ones like CNS lymphoma reflects biases present in the models’ extensive training data, which likely mirror real-world incidence rates, potentially leading to better performance on more frequently encountered pathologies. The limited recognition of IDH status in the differential diagnoses provided by the models is a notable finding. The inability of LLMs to adapt to updated guidelines further complicates their role as standalone diagnostic tools. While we cannot fully explain these phenomena, several potential reasons may exist. These limitations may stem from the probabilistic nature of LLM training, which involves unsupervised learning where parts of the input data are masked and filled with high-probability tokens [17, 26]. # Part 7. Discussion におけるトピック文

トピック文とは、その段落で何を述べるのかを最初に示す文です。

Discussion では、段落ごとに役割が比較的はっきりしています。たとえば、主要結果を述べる段落、結果を解釈する段落、先行研究と比較する段落、限界を述べる段落などです。

そのため、各段落の冒頭にトピック文を置くと、読者は「この段落では何について議論するのか」をすぐに理解できます。

## Discussion でトピック文が重要な理由

Discussion は Results よりも文章の自由度が高く、話が広がりやすいセクションです。トピック文がないと、段落の目的が曖昧になり、読者にとって議論の流れが追いにくくなります。

トピック文には次の役割があります。

- 段落の役割を明確にする
- Results から Discussion への流れを作る
- 先行研究との比較や限界の話に自然につなげる
- 読者が論理の流れを見失わないようにする

Discussion では、1つの段落に1つの中心メッセージを置くのが基本です。その中心メッセージを最初に示すのがトピック文です。


## 段落の役割別トピック文の例

| 段落の役割 | トピック文の例 | その後に書く内容 |
|---|---|---|
| Principal findings | The main finding of this study was that ... | 主要結果を簡潔にまとめる |
| Interpretation | One possible explanation for this finding is that ... | なぜその結果になったかを説明する |
| Comparison | Our findings are consistent with previous studies showing that ... | 先行研究との一致点・相違点を書く |
| Implications | These findings may have clinical implications for ... | 臨床的・研究上の意義を書く |
| Limitations | This study has several limitations. | 限界を具体的に列挙する |
| Conclusion | In conclusion, ... | 研究全体の結論を簡潔に述べる |

例:

```markdown
The main finding of this study was that GPT-4 showed the highest diagnostic accuracy among the evaluated LLMs.

One possible explanation for this result is that structured MRI reports contain standardized descriptions of imaging findings that can be interpreted by LLMs.

This study has several limitations.
```

このように、トピック文を読むだけで Discussion 全体の流れが分かるように書くのが理想です。

# Part 8. Discussion における citation の使い方

Discussion では、Results で得られた自分の結果を解釈し、先行研究や既存の知識と結びつけます。そのため、Introduction と同様に citation が重要です。

ただし、Discussion で citation が必要な場所と不要な場所は区別します。

## Citation が必要な文

以下のような文には citation が必要です。

- 先行研究の結果を述べる文
- 既存のレビューやガイドラインの内容を述べる文
- 自分の結果が先行研究と一致する、または異なると述べる文
- LLM、医療AI、診断基準などについて一般的な知識を述べる文
- 今後の研究の方向性を、既存研究に基づいて述べる文

例:

```markdown
Recent reviews have highlighted both the potential and limitations of LLMs in medical imaging applications [@citation_key].

Our findings are consistent with previous studies showing that LLMs may be useful for generating differential diagnoses from structured reports [@citation_key_1; @citation_key_2].

The WHO classification of central nervous system tumors has been updated to incorporate molecular information such as IDH status [@citation_key].
```

## Citation が不要な文

以下のような文には、通常 citation は不要です。

- 本研究で得られた結果を述べる文
- 自分の Results から直接導かれる解釈
- 本研究の限界を述べる文
- 本研究の結論を述べる文

例:

```markdown
In this study, GPT-4 showed the highest accuracy among the evaluated LLMs.

This study has several limitations, including its single-institution design and relatively small sample size.

In conclusion, LLMs may serve as assistive tools for differentiating intra-axial primary brain tumors using structured MRI reports.
```


## Markdown で citation を入れる方法

この演習では、Session03 と同じように Markdown / Pandoc の citation 形式を使います。

基本形は次の通りです。

```markdown
先行研究を引用する文 [@citation_key].
```

複数の文献をまとめて引用する場合は、セミコロンで区切ります。

```markdown
Several studies have evaluated LLMs in radiology-related tasks [@citation_key_1; @citation_key_2; @citation_key_3].
```

文中で著者名を主語にする場合は、次のように書けます。

```markdown
Bhayana et al. reported that ChatGPT achieved moderate accuracy on radiology board-style questions [@bhayanaPerformanceChatGPT2023].
```

`citation_key` は、必ず `references.bib` に存在するものを使います。Zotero から BibTeX を出力した場合、各文献に対応する citation key を確認してください。


## Discussion で citation を入れる位置の例

以下は、Discussion の各段落で citation を入れる位置の例です。

```markdown
# Discussion

The main finding of this study was that GPT-4 achieved the highest accuracy among the evaluated LLMs.

One possible explanation is that structured MRI reports contain standardized descriptions of imaging findings, which may allow LLMs to identify diagnostically relevant features. Structured reporting has been reported to improve the documentation of imaging findings in neuroradiology [@citation_key].

Our findings are consistent with previous studies suggesting that LLMs can generate differential diagnoses from radiology reports or structured clinical information [@citation_key_1; @citation_key_2]. However, previous reviews have also emphasized that LLMs have important limitations in diagnostic applications [@citation_key_3].

This study has several limitations. First, it was conducted at a single institution with a relatively small sample size. Second, the models were evaluated using structured MRI reports rather than raw imaging data.

In conclusion, LLMs may serve as assistive tools for differentiating intra-axial primary brain tumors using structured MRI reports, although their performance remains inferior to that of board-certified radiologists interpreting actual images.
```

ポイントは、**自分の結果には citation を付けず、先行研究・既存知識・ガイドラインに基づく文には citation を付ける**ことです。

# Part 14. Abstract は最後に書く

Abstract は論文全体の要約です。

多くの自然科学系論文では、次の構造で書きます。

| 項目 | 書く内容 |
|---|---|
| Objective | 研究目的 |
| Materials and methods | 対象、方法、評価指標 |
| Results | 主要な数値結果 |
| Conclusion | 結果から言える結論 |

Abstract は短いですが、論文全体の内容と矛盾してはいけません。

注意点:

- 本文にない結果を書かない
- Results の主要数値を正確に書く
- Conclusion を強くしすぎない
- 略語は必要に応じて初出で説明する
- 最後に本文の Results、Discussion と照合する

Abstract は、Introduction より先に読まれることが多いですが、書くのは本文が固まった後が安全です。

**Objective:**
To evaluate the performance of multiple large language models (LLMs) in the Japanese Diagnostic Radiology Board Examination and compare their accuracy with the performance of human medical students.

Multiple-choice questions from the Japanese Diagnostic Radiology Board Examination were administered to several LLMs, including Gemini 2.5 Pro Preview, GPT-4, and others. **Conclusion:**
Gemini 2.5 Pro Preview demonstrates performance comparable to human examinees in a radiology knowledge assessment. While LLMs show promise as educational and diagnostic support tools, significant performance variability across models and specialties remains a challenge for practical implementation.
