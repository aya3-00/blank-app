# Introduction 本文

> The accurate diagnosis of intra-axial primary brain tumors is a critical yet challenging task in neuroradiology [1, 2]. These tumors, which originate within the brain parenchyma, exhibit a wide range of imaging characteristics that can often overlap, making definitive diagnosis difficult. MRI plays a pivotal role in the evaluation of these lesions, providing detailed information about tumor morphology, composition, and surrounding tissue involvement [3].

脳実質内に発生する原発性脳腫瘍を正確に診断することは、神経放射線診断において非常に重要である一方、難しい課題でもあります [1, 2]。これらの腫瘍は脳実質内に発生し、画像所見が多様で、しばしば互いに重なり合うため、確定的な診断が困難になります。MRI は、これらの病変の評価において中心的な役割を果たし、腫瘍の形態、組成、周囲組織への浸潤や影響について詳細な情報を提供します [3]。

> Recent advancements in artificial intelligence, particularly large language models (LLMs), have shown promise in various medical applications [4,5,6]. These models, trained on vast amounts of textual data, have demonstrated impressive capabilities in natural language understanding and generation. However, current multimodal LLMs, which can process both text and images, still have limited diagnostic capabilities when it comes to direct interpretation of medical images [7, 8]. Given these limitations, structured MRI reports offer a potential alternative for AI-assisted diagnosis. These reports, which contain standardized descriptions of imaging findings, could serve as a proxy for actual image interpretation. A study on structured reporting in neuroradiology demonstrated that this approach ensures reliable detection and documentation of brain pathologies, highlighting its clinical feasibility [9]. Recently, one study has explored the utility of LLMs in providing differential diagnoses based on radiologists’ MRI reports for various common brain tumors, demonstrating promising results [10]. However, this investigation has primarily focused on a broad range of conditions or more common pathologies that are generally easier to differentiate. Furthermore, guidelines for primary brain tumors are frequently updated [11, 12], raising questions about the ability of LLMs to adapt to these evolving diagnostic criteria and classification systems. To date, there has been a lack of studies specifically examining the performance of LLMs in differentiating intra-axial primary brain tumors, which are known for their clinical complexity and diagnostic challenges. Additionally, numerous LLMs, including open-source models, have been recently released to the public [13,14,15,16]. However, their respective capabilities in interpreting structured reports, particularly in the context of brain tumor diagnosis, have not been thoroughly evaluated. This gap in knowledge presents an important area for investigation, as understanding the strengths and limitations of various LLMs could inform their potential integration into clinical workflows.

近年、人工知能、特に大規模言語モデル（large language models: LLMs）の進歩により、さまざまな医療応用への可能性が示されています [4,5,6]。これらのモデルは大量のテキストデータで学習されており、自然言語の理解と生成において優れた能力を示しています。しかし、テキストと画像の両方を処理できる現在の multimodal LLM であっても、医用画像を直接解釈して診断する能力にはまだ限界があります [7, 8]。このような限界を踏まえると、構造化MRIレポートは、AI支援診断のための代替手段となる可能性があります。構造化MRIレポートには、画像所見が標準化された形で記載されているため、実際の画像解釈の代理情報として利用できる可能性があります。神経放射線領域における構造化レポートに関する研究では、この方法が脳病変の信頼性の高い検出と記録を可能にすることが示され、臨床的な実行可能性が示されています [9]。最近、ある研究では、放射線科医によるMRIレポートに基づいて、LLM が一般的な脳腫瘍の鑑別診断を提示する有用性が検討され、有望な結果が示されました [10]。しかし、その研究は主に幅広い疾患、または比較的鑑別しやすい一般的な病変を対象としていました。さらに、原発性脳腫瘍の診療ガイドラインは頻繁に更新されており [11, 12]、変化する診断基準や分類体系に LLM が適応できるかという問題があります。これまで、臨床的に複雑で診断が難しいことが知られている intra-axial primary brain tumors の鑑別において、LLM の性能を特化して検討した研究は不足していました。加えて、open-source model を含む多数の LLM が近年一般に公開されています [13,14,15,16]。しかし、特に脳腫瘍診断の文脈で、構造化レポートを解釈する各モデルの能力は十分に評価されていません。この知識のギャップは重要な検討課題であり、さまざまな LLM の強みと限界を理解することは、将来的な臨床ワークフローへの統合を考える上で有用です。

# Part 1. Introduction の一般的な説明

## Introduction とは

Introduction は、論文の「入口」です。

読者に対して、次のことを説明します。

- なぜこの研究テーマが重要なのか
- これまでの研究で何が分かっているのか
- まだ何が分かっていないのか
- だから本研究では何を調べるのか

Introduction は単なる背景説明ではありません。最後に、本研究の目的へ自然につなげる必要があります。


## Introduction の基本構造

英語論文の Introduction は、多くの場合、次の4段階で書かれます。

| 段落 | 役割 | 書く内容 |
|---|---|---|
| Paragraph 1 | Background | 広い背景。研究テーマがなぜ重要か |
| Paragraph 2 | Previous studies | 先行研究で分かっていること |
| Paragraph 3 | Knowledge gap | まだ十分に分かっていないこと |
| Paragraph 4 | Aim | 本研究の目的 |

この流れは、**背景 -> 先行研究 -> gap -> 目的** と覚えると分かりやすいです。


## Introduction で書かないこと

Introduction では、以下を書きすぎないようにします。

- Results の詳しい数値
- Discussion のような深い解釈
- 自分の感想
- 引用なしの大きな主張
- Methods の細かい手順

悪い例:

> LLMs are very useful and will change medicine.

よい例:

> Large language models (LLMs) have attracted attention as potential tools for medical education and clinical decision support.

引用を入れると、次のようになります。

```markdown
Previous studies have reported that ChatGPT can answer radiology board-style questions with moderate accuracy [@bhayanaPerformanceChatGPT2023].
```


## 引用しなくてもよい文の例

以下のような文は、本研究の目的や自分の研究内容なので、通常は引用不要です。

```markdown
This study aimed to compare the accuracy of multiple LLMs in answering medical imaging-related questions.

We evaluated model performance across radiological subspecialties.
```

ただし、「なぜその目的が重要か」を説明する背景文には引用が必要です。

## Zotero とは

Zotero は、論文や書籍などの文献情報を管理するためのツールです。

Zotero を使うと、以下ができます。

- 論文情報を保存する
- PDF を管理する
- BibTeX 形式で文献情報を書き出す
- Markdown / Pandoc / Word で引用文献リストを作る

この演習では、Zotero に登録した文献を BibTeX 形式で出力し、`references.bib` として使います。


## 文献を探す場所

- PubMed
- Google Scholar
- journal page（出版社ページ）
- DOI
- 論文タイトルでの検索

# Appendix. 英語論文 Introduction の基本

## 1. Introduction の役割

Introduction は、読者に「なぜこの研究を行う必要があるのか」を説明する部分です。

単なる背景説明ではなく、最後に本研究の目的へ自然につなげます。

基本構造は次の4段階です。

1. Background: 広い背景
2. Previous studies: 先行研究で分かっていること
3. Knowledge gap: まだ分かっていないこと
4. Aim: 本研究の目的


## 2. よく使う書き出し

### 背景を書く表現

- Recently, large language models (LLMs) have attracted increasing attention in medicine.
- Artificial intelligence has been increasingly applied to medical education and clinical decision support.
- Radiology is one of the fields in which AI-based tools are actively being investigated.

### 先行研究を書く表現

- Previous studies have reported that LLMs can answer medical examination questions with moderate to high accuracy.
- Several studies have evaluated the performance of ChatGPT in radiology board-style examinations.
- Prior research has suggested that LLM performance may vary depending on question type and domain.

### 未解決の問題を書く表現

- However, the performance of different LLMs across radiological subspecialties remains unclear.
- It is still uncertain whether reasoning-oriented models improve accuracy in medical imaging-related questions.
- Few studies have compared multiple contemporary LLMs using the same set of medical imaging questions.

### 研究目的を書く表現

- Therefore, this study aimed to compare the accuracy of multiple LLMs in answering medical imaging-related questions.
- The purpose of this study was to evaluate model performance across different radiological subspecialties.
- We investigated whether model accuracy differed according to subspecialty and model type.


## 3. 時制の基本

### 一般的な事実・背景

現在形を使います。

- LLMs are increasingly used in medical education.
- Radiology requires accurate interpretation of complex imaging findings.

### 先行研究

現在完了形または過去形を使います。

- Previous studies have evaluated the performance of ChatGPT.
- Bhayana et al. reported that ChatGPT achieved moderate accuracy on radiology board-style questions.

### 本研究の目的

過去形または現在形を使います。

- This study aimed to compare the accuracy of multiple LLMs.
- The purpose of this study is to evaluate the performance of LLMs.


## 4. 4段落テンプレート

以下は、今回のテーマに合わせた最小限のテンプレートです。

```markdown
# Introduction

Large language models (LLMs) have attracted increasing attention in medicine, including medical education, patient communication, and clinical decision support. In radiology, LLMs may be useful for answering knowledge-based questions and supporting learning.

Previous studies have evaluated the ability of LLMs to answer medical and radiology-related examination questions. These studies suggest that LLMs can achieve moderate to high accuracy, although their performance depends on the type and complexity of the questions.

However, it remains unclear how different LLMs perform on medical imaging-related questions across radiological subspecialties. In addition, the effect of reasoning-oriented or thinking modes has not been fully evaluated.

Therefore, this study aimed to compare the accuracy of multiple LLMs in answering medical imaging-related questions. We also evaluated model performance according to radiological subspecialty.
```


## 6. Pandoc の reference DOCX を使う

Pandoc で Markdown から Word ファイルを作成するとき、`--reference-doc=` を使うと、見出し、本文、引用文献リストなどの Word スタイルを指定できます。

たとえば、あらかじめ整えた Word ファイル `Reference.docx` を用意しておくと、そのスタイルを使って `paper.docx` を作成できます。

```bash
pandoc paper.md --citeproc --bibliography=references.bib --reference-doc=Reference.docx -o paper.docx
```

各ファイルの役割は次の通りです。

| ファイル | 役割 |
|---|---|
| `paper.md` | Introduction などを書いた本文ファイル |
| `references.bib` | Zotero から書き出した BibTeX 文献ファイル |
| `Reference.docx` | Word のスタイルを指定するための見本ファイル |
| `paper.docx` | Pandoc で作成される提出用 Word ファイル |

ファイルは、まず同じフォルダに置くと分かりやすいです。

```text
notebooks/
├── paper.md
├── references.bib
├── Reference.docx
└── paper.docx
```

`Reference.docx` は本文の内容を入れるファイルではなく、Pandoc が Word スタイルを読み取るためのテンプレートです。通常は、見出し、本文、図表キャプション、参考文献などのスタイルを Word 側で調整したものを使います。

注意点:

- `Reference.docx` のファイル名とパスを正しく指定する。
- `references.bib` と本文中の `[@citation_key]` が対応していることを確認する。
- `--citeproc` を付けないと、citation が文献リストに変換されない。
- 変換後の `paper.docx` は Word で開き、見出し、本文、文献リストの体裁を確認する。

## PubMed 検索の基本

1. まず広めのキーワードで検索する。
   - 例: `large language model radiology`
2. 検索結果が多すぎる場合は、キーワードを追加する。
   - 例: `large language model radiology diagnosis`
3. Review 論文を探したい場合は `review` を追加する。
   - 例: `large language model medical imaging review`
4. 最近の文献を見たい場合は、PubMed の左側メニューで publication date を絞る。
5. 使えそうな文献を開き、Title、Abstract、Journal、Year、DOI を確認する。
6. Zotero Connector で文献を Zotero に保存する。
7. Zotero から `references.bib` を書き出し、Introduction 本文に `[@citation_key]` を追加する。

検索のコツ:

- 1語ではなく、2-4語を組み合わせる。
- `LLM` よりも `large language model` の方が検索しやすいことがあります。
- 疾患名、検査名、タスク名を組み合わせる。
  - 例: `glioblastoma MRI differential diagnosis`
  - 例: `CNS lymphoma MRI differential diagnosis`
- Introduction に使う文献は、個別研究だけでなく review や guideline も有用です。


## ChatGPT などで文献を探す場合の注意

ChatGPT、Gemini、Claude などに文献検索を手伝ってもらうこともできます。ただし、使い方には注意が必要です。

## 無料版と有料版の違い

サービスや契約状況によって異なりますが、一般的には次のような違いがあります。

| 項目 | 無料版で起こりやすいこと | 有料版・上位プランで期待できること |
|---|---|---|
| 最新情報 | 最新文献にアクセスできない場合がある | Web検索や最新モデルが使える場合がある |
| 検索機能 | 実際のWeb検索ができない場合がある | Web検索機能が使える場合がある |
| 回答の正確性 | もっともらしいが存在しない文献を出すことがある | 改善されることはあるが、確認は必須 |
| 使用回数 | 回数制限がある場合がある | 上限が増える場合がある |
| PDF読解 | 制限がある場合がある | PDFアップロードや長文読解が使える場合がある |

重要なのは、**有料版であっても、文献情報を必ず PubMed、出版社ページ、DOI、Google Scholar などで確認する**ことです。AI が存在しない論文、間違った DOI、誤った著者名を出すことがあります。
